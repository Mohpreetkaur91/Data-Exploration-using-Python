```python
from sklearn import preprocessing
def convert(data):
    number = preprocessing.LabelEncoder()
    for col_name in data.columns:
        data[col_name] = number.fit_transform(data[col_name])
        data=data.fillna(-999)
    return data
```

```python
dataset = convert(ds)
dataset.head(100)
```

```python
X = dataset.drop('class', axis=1)
Y = dataset['class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.10)
```

```python
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
```

#Graph 1: Non-linear Fit of Blood Pressure Over Age



```python
diabetes_data = pd.read_csv('Diabetes.csv')
sns.lmplot(x='Age', y='Diastolic BP', data=diabetes_data, order=2, aspect=2, height=6)
plt.title('Non-linear Trend of Diastolic Blood Pressure Over Time')
plt.xlabel('Time (Days)')
plt.ylabel('Diastolic Blood Pressure')



```

#Graph 2: Risk Factor Scoring System Visualization







```python

sns.barplot(x='class', y='BMI', data=diabetes_data)
plt.title('Average Risk Score by Diabetes Class')
plt.xlabel('Diabetes Class')
plt.ylabel('Risk Score')
```

#Graph 3: Time-series Analysis of Blood Pressure Readings

```python

sns.lineplot(x='Age', y='Diastolic BP', data=diabetes_data)
plt.title('Blood Pressure Readings Over Time')
plt.xlabel('Time')
plt.ylabel('Diastolic Blood Pressure')
```

#Graph 4: Distribution of Diastolic Blood Pressure (Diastolic BP)

```python
plt.figure(figsize=(10, 6))
sns.histplot(diabetes_data['Diastolic BP'], kde=True, color='navy')
plt.title('Distribution of Diastolic Blood Pressure')
plt.xlabel('Diastolic Blood Pressure')
plt.ylabel('Frequency')
plt.show()
```

#Graph 5: Relationship Between BMI and PG Concentration

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='BMI', y='PG Concentration', hue='class', data=diabetes_data, palette='coolwarm')
plt.title('BMI vs. PG Concentration by Diabetes Class')
plt.xlabel('BMI')
plt.ylabel('PG Concentration')
plt.legend(title='Diabetes Class')
plt.show()
```

#Graph 6: Average Insulin Levels Across Different Ages

```python

age_insulin_avg = diabetes_data.groupby('Age')['Serum Ins'].mean().reset_index()

plt.figure(figsize=(12, 7))
sns.lineplot(x='Age', y='Serum Ins', data=age_insulin_avg, marker='o', color='green')
plt.title('Average Insulin Levels Across Different Ages')
plt.xlabel('Age')
plt.ylabel('Average Serum Insulin')
plt.show()
```

#Graph 7: Correlation Heatmap Including Age, Insulin, BMI, and Blood Pressure

```python

cols_for_correlation = ['Age', 'Serum Ins', 'BMI', 'Diastolic BP', 'PG Concentration']
correlation_matrix = diabetes_data[cols_for_correlation].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='cividis')
plt.title('Correlation Heatmap of Selected Variables')
plt.show()
```
